# Who had the most Entry Frags in the Tournament?

```python
Players= df['Players'].head(20)
First_Bloods= df['First_Bloods'].head(20)

fig= plt.figure(figsize =(25,15))

plt.bar(Players,First_Bloods)

plt.show()
```

# Who had the most First Deaths in the Tournament?

```python
Players= df['Players'].head(20)
First_Deaths= df['First_Deaths'].head(20)

fig= plt.figure(figsize =(25,15))

plt.bar(Players,First_Deaths, color='Blue')

plt.show()
```

# Players with the Highest/Lowest K/D Ratio

```python
sns.lineplot(x=df['Players'].head(20),y=df['KD_Ratio'].head(20), color = "y")
plt.xticks(rotation = 65)
plt.xlabel('Players')
plt.ylabel('KD_Ratio')
plt.show()         
```

```python
sns.lineplot(x=df['Players'].tail(20),y=df['KD_Ratio'].tail(20), color = 'g')
plt.xticks(rotation = 65)
plt.xlabel('Players')
plt.ylabel('KD_Ratio')
plt.show()
```

# Headshot Percentage Accuracy amongst Team

```python
df.plot(kind='scatter',
        x='Teams',
        y= 'Headshot_Percentage',
       title= 'Headshot Percentage for Teams')
plt.show()
```

# Teams' Impact Per Round

```python
sns.pairplot(df, vars=['Kills_PR','Assists_PR','FirstBlood_PR','Average_Damage_PR'],
            hue= 'Teams')
plt.show()
```
